```python
import time

def print_one_line(lang_file):
    time.sleep(0.2)
    line = lang_file.readline()
    if line:
        print(line)
    else:
        lang_file.seek(0)
    print_one_line(lang_file)
    
lang_file = open("languages.txt")
print_one_line(lang_file)
```
